```python
import csv
import matplotlib.pyplot as plt
```

```python
data = []
with open('pose_keypoints_data_20-24.csv', mode='r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        data.append(row)
```

```python
average_keypoints_per_frame = [float(row[2]) for row in data]
total_keypoints = [int(row[1]) for row in data]
```

```python
plt.hist(average_keypoints_per_frame, bins=20, color='blue', alpha=0.7)
plt.xlabel('Среднее количество точек на кадр')
plt.ylabel('Частота')
plt.title('Гистограмма среднего количества точек на кадр во всех видео')
plt.show()
```

```python
plt.hist(total_keypoints, bins=20, color='green', alpha=0.7)
plt.xlabel('Общее количество точек')
plt.ylabel('Частота')
plt.title('Гистограмма общего количества точек во всех видео')
plt.show()
```
